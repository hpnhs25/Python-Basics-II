```python

# Import pathlib from CSV library
from pathlib import Path
import csv

# Set the file path
csvpath = Path("./Resources/budget_data.csv")

# Initial List Set-Ups
month_change_list = []
net_change_list = []
greatest_increase = ["ex", 0]
greatest_decrease = ["ex", 100000000000]

gri_month = []
gri_increase = []

grd_month = []
grd_decrease = []

# Initial Variables
total_net = 0
total_month = 0

# Open csv file as an object
with open(csvpath, 'r') as csvfile:

    # Pass csv file to the csv reader
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # Read Header Row and First Row
    header = next(csvreader)
    firstrow = next(csvreader)
    total_month += 1
    
    # Variables for profit change calculation
    total_net = total_net + int(firstrow[1])
    prev_net = int(firstrow[1])
    
    # For Loop to complete calculations
    for row in csvreader:
        # print row
        
        total_month += 1 # Total Month counter
        
        # Month to month profit calculations
        total_net = total_net + int(row[1])
        net_change = int(row[1]) - prev_net
        prev_net = int(row[1])
        
        # Lists for monthly profit change and specific month
        net_change_list.append(net_change)
        month_change_list.append(row[0])
        
        # Find greatest increase and decrease in month to month profit change
        # Isolate the date and the amount
        if net_change > greatest_increase[1]:
            greatest_increase[1] = net_change
            greatest_increase[0] = row[0]
        if net_change < greatest_decrease[1]:
            greatest_decrease[1] = net_change
            greatest_decrease[0] = row[0]
    
    # Average change calculation rounded
    net_monthly_avg = sum(net_change_list) / len(net_change_list)
    net_monthly_avg_rounded = round(net_monthly_avg, 2)
    
    # Split the month and $ Amounts out for report formatting
    gri_month = greatest_increase[0]
    gri_increase = greatest_increase[1] 
    
    grd_month = greatest_decrease[0]
    grd_decrease = greatest_decrease[1]
    
    # Print Financial Analysis Summary
    print(f"Financial Analysis")
    print(f"------------------------------------------------")
    print(f"Total Months: {total_month}")
    print(f"Total: ${total_net}")
    print(f"Average Change: ${net_monthly_avg_rounded}")
    print(f"Greatest Increase in Profits: {gri_month} ${gri_increase}")
    print(f"Greatest Decrease in Profits: {grd_month} ${grd_decrease}")

    # Get Output File Ready
    header = ["Total Months", "Total", "Average Change", "Greatest Incr Month", "Greatest Incr Profit", "Greatest Decr Month", "Greatest Decr Profit"]
    metrics = [total_month, total_net, net_monthly_avg_rounded, gri_month, gri_increase, grd_month, grd_decrease]
    
    # Create output file
    output_path = Path("./Resources/PyBank.csv")
    with open(output_path, "w") as csvfile:
        csvwriter = csv.writer(csvfile, delimiter=",")
        csvwriter.writerow(header)
        csvwriter.writerow(metrics)
    
    
```
