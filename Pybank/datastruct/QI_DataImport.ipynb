```python

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(14, 5)) # Customize plot size

plt.plot(appl, 'b') # blue line color
plt.plot(appl, 'ro') # red dots
plt.grid(True) # grid layout to plot
plt.title('Apple Close Price Graph') # title to plot
plt.xlabel('Trading Day') # x-axis label
plt.ylabel('Apple Closing Price') # y-axis label
plt.show() # show visual of plot


```

```python

appl = data[['Close','Open']]
```

```python

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(14, 5)) # Customize plot size

plt.plot(appl["Close"], lw=1.5, label='Close')
plt.plot(appl["Open"], lw=1.5, label='Open')

plt.plot(appl, 'ro') # red dots

plt.grid(True) # grid layout to plot

plt.legend(loc=0)

plt.axis('tight')

plt.title('Apple Price Graph') # title to plot
plt.xlabel('Time') # x-axis label
plt.ylabel('Index') # y-axis label
plt.show() # show visual of plot


```

```python
import numpy as np

y = np.random.standard_normal((100, 2)) # Random Data Created

plt.figure(figsize=(7, 5))
plt.scatter(y[:, 0], y[:, 1], marker='o')

plt.grid(True)
plt.xlabel('1st dataset')
plt.ylabel('2nd dataset')
plt.title('Scatter Plot')
plt.show()
```

```python
np.random.seed(100)
y = np.random.standard_normal(size=1000)
plt.figure(figsize=(10, 5))
plt.hist(y, label=['Return Distribution'])

plt.grid(True)
plt.legend(loc=0)
plt.ylabel('Frequency')
plt.xlabel('Return in %')
plt.title('Histogram')
plt.show()


```
